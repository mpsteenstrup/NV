```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from urllib.request import urlopen
```

```python
# URL of the image
url = 'https://github.com/mpsteenstrup/Billedmanipulation/blob/master/billeder/mps.jpg?raw=true'
# Step 1: Fetch the image from the URL
with urlopen(url) as response:
    # Read the image from the response
        image = mpimg.imread(response, format='JPG')

        # Step 2: Extract the red channel
        # The red channel is the first channel (index 0).
        red_channel = image[:, :, 0]

        # The red channel image should have the same dimensions as the original.
        # Create an empty image with all zeros (black) with the same size as the original.
        img = np.zeros_like(image)

        # Assign the red channel of the original image to the corresponding channel of the new image.
        img[:, :,0] = red_channel

        # Step 3: Display the images
        # Display original image
        plt.subplot(1, 2, 1)
        #plt.imshow(image)
        plt.imshow(image)
        plt.title('Original Image')
        plt.axis('off')

        # Display the image with only the red channel
        plt.subplot(1, 2, 2)
    #    plt.imshow(img)
        plt.imshow(img)
        plt.title('manipulated')
        plt.axis('off')
        
        plt.show()
```

```python
# grayscale
# URL of the image
url = 'https://github.com/mpsteenstrup/Billedmanipulation/blob/master/billeder/mps.jpg?raw=true'
# Step 1: Fetch the image from the URL
with urlopen(url) as response:
    # Read the image from the response
        image = mpimg.imread(response, format='JPG')

        img_gennemsnit = image[:,:,0]*0.333+image[:,:,1]*0.333+image[:,:,2]*0.333
        img_vaegtet_gennemsnit = 0.2989 * image[:, :, 0] + 0.5870 * image[:, :, 1] + 0.1140 * image[:, :, 2]
        # Step 3: Display the images
        # Display original image
        plt.subplot(1, 3, 1)
        #plt.imshow(img_gennemsnit)
        plt.imshow(img_gennemsnit, cmap='gray')
        plt.title('gennemsnit')
        plt.axis('off')

        # Display the image with only the red channel
        plt.subplot(1, 3, 2)
    #    plt.imshow(img_vaegtet)
        plt.imshow(img_vaegtet_gennemsnit, cmap='gray')
        plt.title('vægtet gennemsnit')
        plt.axis('off')


        plt.subplot(1,3,3)
        plt.imshow(img_gennemsnit-img_vaegtet_gennemsnit, cmap='gray')
        plt.title('forskel')
        plt.axis('off')
    
        plt.show()
```
